```python
from google.colab import files
uploaded = files.upload()

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

file_name = list(uploaded.keys())[0]
print("Uploaded file:", file_name)

df = pd.read_csv(file_name)

print("Dataset:")
print(df.head())

print("\nMissing Values:")
print(df.isnull().sum())

df = df.dropna()

df["Total"] = df["Maths"] + df["Science"] + df["English"]
df["Average"] = df["Total"] / 3

print("\nUpdated Dataset:")
print(df.head())

math_avg = np.mean(df["Maths"])
science_avg = np.mean(df["Science"])
english_avg = np.mean(df["English"])

print("\nAverage Marks:")
print("Maths:", math_avg)
print("Science:", science_avg)
print("English:", english_avg)

top_student = df.loc[df["Total"].idxmax()]

print("\nTop Performer:")
print(top_student["Name"])
print("Total Marks:", top_student["Total"])

highest_subject = max(
    {
        "Maths": math_avg,
        "Science": science_avg,
        "English": english_avg
    },
    key=lambda x: {
        "Maths": math_avg,
        "Science": science_avg,
        "English": english_avg
    }[x]
)

print("\nHighest Scoring Subject:", highest_subject)

plt.bar(
    ["Maths", "Science", "English"],
    [math_avg, science_avg, english_avg]
)
plt.title("Average Marks in Each Subject")
plt.xlabel("Subjects")
plt.ylabel("Average Marks")
plt.show()

plt.hist(df["Total"], bins=10)
plt.title("Distribution of Total Marks")
plt.xlabel("Total Marks")
plt.ylabel("Number of Students")
plt.show()
```
